```python
import pandas as pd
import datetime
# Extract labels for the same month as current month
def find_current_labels(labels_df):
    mask = []
    for row in labels_df.iterrows():
        start_month = datetime.datetime.strptime(row[1]['start'], tm_format).month
        end_month = datetime.datetime.strptime(row[1]['end'], tm_format).month
        if start_month >= current_month >= end_month:
            mask.append(True)
        else:
            mask.append(False)
    return labels_df.loc[mask,:]
```

```python
current_labels_df = find_current_labels(labels_df)
```
